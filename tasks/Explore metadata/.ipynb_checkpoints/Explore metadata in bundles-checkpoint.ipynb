This metadata vignette lays out how to traverse the metadata in a set of data bundles downloaded from the HCA DSS.

Fields of interest we will extract include
- IDs for each entity
- Donor
    - species
    - sex
    - development stage
    - diseases
- Tissue, cell line or organoid
    - organ/organ part/ model organ
- Protocols
    - library construction method
    - sequencing method
    - cell count

```python
import os
import json
from pprint import pprint
 
rootDir = '.'
for dirName, subdirList, fileList in os.walk(rootDir):
    if dirName != rootDir:
        print('Found directory: %s' % dirName)
        for fname in fileList:
            if 'donor_organism' in fname:
                print('\t%s' % fname)
                with open(dirName+'/'+fname) as f:
                    data = json.load(f)
                    donor_id = data['biomaterial_core']['biomaterial_id']
                    diseases = data['diseases']
                    species = data['genus_species']
                    print(donor_id + "\t" + str(diseases) + "\t" + str(species))
```
